```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Kepler90Planets is the simple planets data with scaling planets size.
img = cv2.imread("Kepler90planets.jpg")
```

##Data Preprocessing

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
(thresh, output2) = cv2.threshold(gray_image, 120, 255, cv2.THRESH_BINARY)
output2 = cv2.GaussianBlur(output2, (5, 5), 1)
output2 = cv2.Canny(output2, 180, 255)
plt.imshow(output2, cmap = plt.get_cmap("gray"))
circles = cv2.HoughCircles(output2,cv2.HOUGH_GRADIENT,1,10,param1=180,param2=27,minRadius=20,maxRadius=60)
circles = np.uint16(np.around(circles))
for i in circles[0,:]:
    # draw the outer circle
    cv2.circle(img,(i[0],i[1]),i[2],(0,255,0),2)
    # draw the center of the circle
    cv2.circle(img,(i[0],i[1]),2,(0,0,255),3)
    
plt.imshow(img)
```

```python
data_img = cv2.imread("Kepler90planets.jpg")
gray = cv2.cvtColor(data_img, cv2.COLOR_BGR2GRAY)
data_img = cv2.medianBlur(gray, 5)
circles = cv2.HoughCircles(data_img,cv2.HOUGH_GRADIENT,1,18, param1=100,param2=30,minRadius=20,maxRadius=60)
circles = np.uint16(np.around(circles))
for i in circles[0,:]:
    # outer circle
    cv2.circle(data_img,(i[0],i[1]),i[2],(0,255,0),2)
    # center of circle
    cv2.circle(data_img,(i[0],i[1]),2,(0,0,255),3)
    
#cv2.imshow("HoughCirlces",	data_img)
#cv2.waitKey()
#cv2.destroyAllWindows()
plt.imshow(data_img)
```
